## CODING TASK #4. PLOT HISTOGRAMS USING MATPLOTLIB

```python
# A histogram represents data using bars with various heights
# Each bar groups numbers into specific ranges
# Taller bars show that more data falls within that specific range

daily_return_df = pd.read_csv('crypto_daily_returns.csv')
daily_return_df
```

```python
mu = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
 
num_bins = 40

# Plot the histogram of the returns
plt.figure(figsize = (20, 15)) 

daily_return_df['BTC'].plot.hist(bins = num_bins, color = 'b', alpha = 0.7, figsize = (12, 6))

plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: mu=' + str(mu) + ', sigma=' +str(sigma));
 
```

**PRACTICE OPPORTUNITY #4 [OPTIONAL]:**
- **Plot the histogram for BTC and ETH returns using 60 bins. Display all histograms on the same figure**


```python
mu_BTC = round(daily_return_df['BTC'].mean(), 2) # mean of distribution
sigma_BTC = round(daily_return_df['BTC'].std(), 2) # standard deviation of distribution
 
mu_ETH = round(daily_return_df['ETH'].mean(), 2) # mean of distribution
sigma_ETH = round(daily_return_df['ETH'].std(), 2) # standard deviation of distribution

num_bins = 60

# Plot the histogram of the returns
plt.figure(figsize = (20, 15)) 

daily_return_df[['BTC','ETH']].plot.hist(bins = num_bins, alpha = 0.7, figsize = (12, 6))

plt.grid()
plt.ylabel('Probability')
plt.title('Histogram: BTC mu=' + str(mu_BTC) + ', BTC sigma=' +str(sigma_BTC) + ' , ETH mu=' + str(mu_ETH) + ', ETH sigma=' +str(sigma_ETH));
```

## CODING TASK #6. PLOT PAIRPLOT, DISPLOT, AND HEATMAPS/CORRELATIONS USING SEABORN

```python
# Plot the pairplot
sns.pairplot(cancer_df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )

#  This is cool, and notice that on the diagonal, we see the distribution of each individual variable, e.g. mean radius, for observations with target value = 1 versus 0
```

```python
# Strong correlation between the mean radius and mean perimeter, mean area and mean primeter (both 0.99)
plt.figure(figsize = (20, 10)) 
sns.heatmap(cancer_df.corr(), annot = True) 
```

```python
# plot the distplot 
# Displot combines matplotlib histogram function with kdeplot() (Kernel density estimate)
# KDE is used to plot the Probability Density of a continuous variable. 

sns.distplot(x = cancer_df['mean radius'], bins = 15, color = 'blue')
```
